# knn algorithm

```python
def eucleoid_dist(p1,p2):
    return np.sum((p2-p1)**2)**0.5
```

```python
def knn(x,y,test,k=11):
    d=[]
    for i in range(x.shape[0]):
        dist=eucleoid_dist(test,x[i])
        d.append([dist,y[i]])
    d=np.array(sorted(d,key=lambda x:x[0]))[:,1]
    d=d[:k]
    print(d)
    p=np.unique(d,return_counts=True)
    max_idx=np.argmax(p[1])
    pred=p[0][max_idx]
    
    print("the predicted value is:" ,end="")
    print(pred)
    return pred
    
```

# testing the algorithm for 10 values

```python
def drawing(samp):
    plt.imshow(samp.reshape((28,28)),cmap='gray')
    plt.show()
```

```python
for j in range(10,20):
    plt.subplot(2,5,j-9)
    knn(pixels,label,test_pixels[j],11)
    drawing(test_pixels[j])
    print()
    
```

## CALCULATION  OF ACCURACY
### Using hold out validation

```python
split=int(0.7*data.shape[0])      #dividing the data into 70-30 ratio for training and testing respectively
data_train=data[:split,:]
data_test=data[split:,:]
```

```python
print(data_train.shape)
print(data_test.shape)
```

```python
predicted_value=[]
for i in range(data_test.shape[0]):
    p=knn(data_train[:,1:],data_train[:,0],data_test[i,1:],11)
    predicted_value.append(p) 
```

```python
#converting into integer labels
y_pred=[]
for i in predicted_value:
    y_pred.append(int(i))
```

```python
y_true=data_test[:len(predicted_value),0]
```

```python
accuracy=np.sum(y_pred==y_true)/len(y_true)*100
```

```python
print("ACCURACY = ",accuracy)
```
